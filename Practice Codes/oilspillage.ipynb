```python
import os

results_dir = '/content/runs/segment/train4'

if os.path.exists(results_dir):
    print(f"Contents of '{results_dir}':")
    for item in os.listdir(results_dir):
        print(f"- {item}")
else:
    print(f"Error: The directory '{results_dir}' does not exist.")
```

```python
import shutil
import os

results_dir = '/content/runs/segment/train4'
output_filename = '/content/train4'

if os.path.exists(results_dir):
    print(f"Compressing '{results_dir}' into '{output_filename}.zip'...")
    shutil.make_archive(output_filename, 'zip', results_dir)
    print(f"Compression complete. You can download '{output_filename}.zip' from the Colab file browser.")
else:
    print(f"Error: The directory '{results_dir}' does not exist and cannot be compressed.")
```
